```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = load_iris()
```

```python
data
```

```python
data.DESCR
```

```python
data.data
```

```python
data.feature_names
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data.data , columns = data.feature_names)
```

```python
df
```

```python
df['target'] = data.target
```

```python
df
```

```python
df.target.unique()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state= 1, test_size= 0.2)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier( criterion= 'entropy', splitter= 'best', min_samples_split = 2)
```

```python
model
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize= (10, 8))
tree.plot_tree(model, filled= True, rounded= True, node_ids= True, max_depth= 10)
plt.show()
```

```python
classifier = DecisionTreeClassifier(criterion='gini')
```

```python
classifier.fit(X_train, Y_train)
```

```python
plt.figure(figsize = (8,8))
tree.plot_tree(classifier, filled= True, fontsize= 8, impurity=True, max_depth= 4)
plt.show()
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred1 = classifier.predict(X_test)
```

```python
print(y_pred)
print(y_pred1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, y_pred)
```

```python
accuracy_score(Y_test, y_pred1)
```

```python
#Post Pruning>> After making complete decision tree you have to cut it to avoid overfitting
```

```python
#Ovservation>> Observe the decision tree plot, You will realisise the last split is actually overfitting
#diving the data in such a way that each node is 100 percent pure leading to low no of data point, i.e : 1 and 3
#So Reduce the depth of Decision Tree
#It is used for small dataset
```

```python
classifier1 = DecisionTreeClassifier(criterion= 'entropy', max_depth= 3)
classifier1.fit(X_train, Y_train)
```

```python
tree.plot_tree(classifier1, filled= True, node_ids= True)
plt.show()
```

```python
tree.plot_tree(classifier, filled=True, node_ids= True)
plt.show()
```

```python
y_pred2 = classifier1.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred2)
```

```python
#The Accuracy is same, earlier it was a overfitting model
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,y_pred2))
```

```python
#Pre_Pruning Using HYperparameter>> 

Parameter = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,6,7,8,9],
    'max_features':['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = DecisionTreeClassifier()
```

```python
model1 = GridSearchCV(clf, param_grid = Parameter, cv = 5, scoring = 'accuracy', verbose = 3)
```

```python
model1.fit(X_train, Y_train)
```

```python
y_pred4 = model1.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred4)
```

```python
model1.best_score_
```

```python
model1.best_estimator_.predict(X_test)
```

```python
print(classification_report(Y_test, y_pred4))
```
