##1. Write a Python program to train a Decision Tree Classifier on the iris dataset and print the model accuracy.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Target'] = data.target
```

```python
df.sample(10)
```

```python
df.Target.unique()
```

```python
#Splitting the values in X and Y

X = df.drop('Target', axis=1)
y = df['Target']
```

```python
print(X.shape)
print(y.shape)
```

```python
#Splitting the x and y into Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state = 1, test_size = 0.2)
```

```python
#Training the decision Tree Model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
#Printing the model Accuracy
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy_score(Y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, y_pred)
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(model, filled=True,max_depth= 4 )
plt.show()
```
