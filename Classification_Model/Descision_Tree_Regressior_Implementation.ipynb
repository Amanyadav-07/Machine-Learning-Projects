```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data.data, columns= data.feature_names)
```

```python
df['Price'] = data.target
```

```python
print(df)
```

```python
df.drop('Target', axis = 1, inplace= True)
```

```python
df
```

```python
df.sample(10)
```

```python
#Seperate Independent Variable and Dependent Variable
```

```python
X = df.iloc[:, :-1]
Y= df.iloc[:, -1]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state= 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred, Y_test)
```

```python
#Hyperparameter Tuning
```

```python
params = {
    'criterion' : ['squared_error','absolute_error', 'friedman_mse'],
    'splitter': ['best', 'random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressior = DecisionTreeRegressor()
```

```python
model1  = GridSearchCV(regressior, param_grid= params, cv = 2, scoring= 'neg_mean_squared_error', verbose = 2) 
```

```python
model1.fit(X_train, Y_train)
```

```python
model1.best_params_
```

```python
from sklearn import tree
```

```python
plt.figure(figsize= (10, 8))
tree.plot_tree(model1, filled= True)
plt.show()
```
