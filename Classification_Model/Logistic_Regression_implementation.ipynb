```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
print(data.DESCR)
```

```python
dataframe = pd.DataFrame(data.data, columns = data.feature_names)
dataframe['Target'] = data.target
```

```python
dataframe.columns
```

```python
dataframe.Target.unique()
```

```python
df = dataframe[dataframe.Target!= 2]
```

```python
df.Target.unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:, -1]
```

```python
X_Train, X_test, Y_Train, Y_Test = train_test_split(X, Y, random_state = 1, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model
```

```python
model.fit(X_Train, Y_Train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.coef_
```

```python
model.predict_proba(X_test)  #Corrospoding to higher probablity class is predicted, The cuttof is 0.5
```

```python
#Evaluation Matrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(Y_Test, y_pred)
```

```python
print(classification_report(Y_Test, y_pred))
```

```python
#First Two rows 0 and 1>> precision, recall, f1_score, and no of datapoints with respect to respective class
#Macro Avg:- For each class take simple avg>> so Avg precision of class 0 and class 1 will be macro avg
#similarly for recall, f1_Score
#Support will be total no of datapoints

#Weighted_avg is used in case of class imbalance, it gives importance to performance of dominant class
```

```python
from sklearn.metrics import roc_curve, auc
#to get predicted probablities of class1 
Y_pred_proba = model.predict_proba(X_test)[:,1]
```

```python
#Reciver Operating Charecteristics Curve
#inputs : Y-test, Y_pred_Proba. It will return TPR, FPR, with different cuttof probablities
fpr, tpr, thresholds = roc_curve(Y_Test, Y_pred_proba)
```

```python
print(fpr)
print(tpr)
print(thresholds)
```
