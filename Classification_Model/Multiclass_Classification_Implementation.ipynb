```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_classes= 3, n_samples= 1000, n_features= 10, n_informative= 5, n_redundant= 5, random_state= 1) 
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state= 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#OVR[One Versus Rest]
```

```python
model = LogisticRegression(multi_class= 'ovr' , solver ='lbfgs')
```

```python
model
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from  sklearn.metrics import confusion_matrix, accuracy_score , classification_report
```

```python
print(confusion_matrix(Y_test, y_pred))
print(f'accuracy is {accuracy_score(Y_test, y_pred)}')
print(classification_report(Y_test, y_pred))
```

```python
model1= LogisticRegression(multi_class= 'multinomial', solver= 'lbfgs')
```

```python
model1
```

```python
model1.fit(X_train, Y_train)
```

```python
model1.coef_
```

```python
pred1 = model1.predict(X_test)
```

```python
print(confusion_matrix(Y_test, pred1))
print(f'accuracy is {accuracy_score(Y_test, pred1)}')
print(classification_report(Y_test, pred1))
```
