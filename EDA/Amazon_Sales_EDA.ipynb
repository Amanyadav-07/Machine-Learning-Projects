```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('amazon.csv')
```

```python
df
```

```python
df.drop('product_id', axis=1, inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.rating.unique()
```

```python
df.rating = df.rating.replace({'|' : 0}).astype(float)
```

```python
df.discount_percentage = df.discount_percentage.replace('%', '', regex=True).astype(int)
```

```python
df.actual_price.unique()
```

```python
df.actual_price = df.actual_price.replace('₹', '', regex=True)
```

```python
df.actual_price = df.actual_price.replace(',', '', regex=True).astype(int)
```

```python
df.discounted_price = df.discounted_price.replace('₹', '', regex=True)
```

```python
df.discounted_price = df.discounted_price.replace(',', '', regex=True).astype(float)
```

```python
df.info()
```

```python
df.drop('actual_rice', axis=1, inplace=True)
```

```python
df.drop('img_link', axis=1, inplace=True)
```

```python
df.drop('product_link', axis=1, inplace=True)
```

```python
df.info()
```

1. What is the average rating for each product category?

```python
df.groupby('category')['rating'].mean()
```

2. What is the top rating_count products by category?

```python
df.groupby('rating_count')['category'].apply(lambda x: x.sort_values(ascending=False)).head()
```

3. What is the distribution of discounted prices vs actual prices?

```python
df.info()
```

```python
plt.subplot(1,2,1)
sns.violinplot(df.discounted_price, palette = 'Set3')
plt.title('Distribution of Discounted Prices')

plt.subplot(1,2,2)
sns.violinplot(df.actual_price, palette = 'Set1')
plt.title('Distribution of Actual Prices')

plt.tight_layout()
plt.show()
```

```python
#insights:- both columns have outliers present
```
