```python
# Locate the index of the row with '5m' in 'Duration_hrs'
index_5m = df.index[df.Duration_hrs == '5m'][0]

# Update 'Duration_hrs' and 'Duration_min' for that row
df.loc[index_5m, 'Duration_hrs'] = 0  # Set 'Duration_hrs' to 0
df.loc[index_5m, 'Duration_min'] = 5  # Set 'Duration_min' to 5

# Convert the columns to integers (excluding the row with '5m' which was handled separately)
df.Duration_hrs = pd.to_numeric(df.Duration_hrs, errors='coerce').fillna(0).astype(int)
df.Duration_min = pd.to_numeric(df.Duration_min, errors='coerce').fillna(0).astype(int)
```
