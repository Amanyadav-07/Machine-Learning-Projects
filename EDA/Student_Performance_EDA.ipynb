```python
#outlier detection
#To detect outlier>> dist, boxplot, violinplot
```

```python
student_grades_df = df.copy()
```

```python
student_grades_df.drop(['G1','G2','G3'], axis = 1, inplace = True)
```

```python
student_grades_df
```

```python
#function to detect outliers
def detect_outlier(columns):
  outlier_indices = []
  for column in columns:
    Q1 = student_grades_df[column].quantile(0.25)
    Q3 = student_grades_df[column].quantile(0.75)
    IQR = Q3 - Q1
    #mask is a boolean condition
    mask = (student_grades_df[column] >= Q1 - 1.5*IQR) & (student_grades_df[column] <= Q3 + 1.5*IQR)
    mask = mask.to_numpy() #bool to numpy array

    #find the indices of value that do not satisfy the mask condition that will be outliers

    false_indices = np.argwhere(~mask)
    outlier_indices.append(false_indices)
  return np.unique(np.concatenate(outlier_indices).ravel())

#np.concatinate(outlier_indices)>> combine all the outlier index array into a single array
#so each column will have one array
#.ravel()>> flatten the multidimensional array into 1D array
```

```python
num_cols = ['age','absences','traveltime','studytime']
detect_outlier(num_cols)
```
