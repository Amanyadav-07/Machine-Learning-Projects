```python
df['us_economic_sector']
```

```python
us_economic_counts = {}
for value in df['us_economic_sector'].dropna():
  if value in us_economic_counts:
    us_economic_counts[value] += 1
  else:
    us_economic_counts[value] = 1
```

```python
us_economic_counts
```

```python
#no has key and names od sector as list
usecollabels = []
usecolvalues = []
for key, value in us_economic_counts.items():
  usecollabels.append(key)
  usecolvalues.append(value)
```

```python
plt.figure(figsize=(10, 10))
# Pass usecolvalues as the first argument for wedge sizes
# and usecollabels as labels
plt.pie(usecolvalues[:10], labels=usecollabels[:10], autopct='%1.1f%%', explode = [0.035,0,0,0,0,0,0,0,0,0], shadow =True)
plt.show()
```
