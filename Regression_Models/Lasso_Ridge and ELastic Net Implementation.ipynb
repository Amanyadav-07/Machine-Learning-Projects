```python
#Seperate into x and y
```

```python
x = df.drop('mpg', axis = 1)
y = df['mpg']
```

```python
x
```

```python
y
```

```python
#train-test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
X_train.shape
```

```python
#Simple linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, Y_train)
```

```python
model.coef_
```

```python
for i, col_name in enumerate(X_train.columns):
    print('The coefficient for {} is {}'.format(col_name, model.coef_[i]))
```

```python
#Observations:- Coefficients are retatively smaller, if one independent variable changes
#There will be not much difference in prediction
#This is sometime called as smoother

#These features might not be contributing in model training
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred_linear = model.predict(X_test)
```

```python
r2_linear = r2_score(Y_test, y_pred_linear)
```

```python
print(f"The r2 score for linear model is {r2_linear}")
```

```python
#Regularized Model
#1. Ridge Regression
#in  Practical implementation lambda is alpha
```

```python
from sklearn.linear_model import Ridge
```

```python
Ridge_Regression_model = Ridge(alpha = 0.2)
```

```python
Ridge_Regression_model
```

```python
Ridge_Regression_model.fit(X_train, Y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print('The coefficient for {} is {}'.format(col_name, Ridge_Regression_model.coef_[i]))
```

```python
#For ridge regression evaluation
y_pred_ridge = Ridge_Regression_model.predict(X_test)
r2_ridge = r2_score(Y_test, y_pred_ridge)
```

```python
print(f"The r2 score for ridge model is {r2_ridge}")
```

```python
#observations:- We don't see much variations in coeff of ridge regression as compared to Linear Regression
```

```python
from sklearn.linear_model import Lasso
```

```python
Lasso_Regeession_model = Lasso(alpha = 0.2)
```

```python
Lasso_Regeession_model
```

```python
Lasso_Regeession_model.fit(X_train, Y_train)
```

```python
Lasso_pred = Lasso_Regeession_model.predict(X_test)
```

```python
for i, col_name in enumerate(X_train.columns):
    print('The coefficient for {} is {}'.format(col_name, Lasso_Regeession_model.coef_[i]))
```

```python
#Three features coefficent is 0, Lasso helps in feature selection
```

```python
r2_score_Lasso = r2_score(Y_test, Lasso_pred)
```

```python
print(f'The r2 score for Lasso model is {r2_score_Lasso}')
```

```python
#ELastic Net
from sklearn.linear_model import ElasticNet
```

```python
Elastic_Net_model = ElasticNet(alpha = 0.2, l1_ratio = 0.5)
Elastic_Net_model.fit(X_train, Y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print('The coefficient for {} is {}'.format(col_name, Elastic_Net_model.coef_[i]))
```

```python
r2_score_elastic_net = r2_score(Y_test, Elastic_Net_model.predict(X_test))
```

```python
print(f'The r2 score for Elastic Net model is {r2_score_elastic_net}')
```
