```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.keys()
```

```python
diabetes.feature_names
```

```python
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
df['target'] = diabetes.target
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6']]
y = df['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit_transform(X_train)
```

```python
scaler.transform(X_test)
```

```python
scaler
```

```python
plt.scatter(X_train['bmi'], Y_train, color = 'black', label = 'Actual')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
reg = LR.fit(X_train, Y_train)
```

```python
reg.score(X_train, Y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X_test, Y_test)
```

```python
plt.scatter(X_train['s6'], Y_train, color = 'black', label = 'Actual')
plt.plot(X_train['s6'], reg.predict(X_train), color = 'red', label = 'Predicted')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
#interpretation>> age has coef value -30.62
#1 unit change in age then target value decreases by -30 units on an average when all other features are kept constant.
```
